```python
df['Age'] = df['Age'].astype(int)
df['Survived'] = df['Survived'].astype(bool)
df['Passenger_Class'] = df['Passenger_Class'].astype('category')
```

# ***CAR DATASET (EDA + Visualization)***

```python
import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/imports-85.csv"
df = pd.read_csv(url)

df.head()
```

```python
import numpy as np

df = df.replace('?', np.nan)

# convert important columns
cols = ['price', 'horsepower', 'engine-size', 'city-mpg']
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# drop rows where price is missing
df = df.dropna(subset=['price'])
```

```python
df['make'].value_counts()
```

```python
df['body-style'].value_counts()
```

```python
df['drive-wheels'].value_counts(normalize=True)
```

```python
df['fuel-system'].value_counts()
```

```python
df.groupby('make')['price'].mean().sort_values(ascending=False)
```

```python
df.groupby('body-style')['horsepower'].mean().sort_values(ascending=False)
```

```python
df.groupby('drive-wheels')['city-mpg'].mean().sort_values(ascending=False)
```

```python
df.groupby('body-style')['price'].agg(['min','max','mean'])
```

```python
df.groupby(['make','body-style'])['price'].mean().sort_values(ascending=False)
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
corr.unstack().sort_values(ascending=False)
```

```python
corr['price'].sort_values(ascending=False)
```

```python
corr['price']['horsepower']
```

```python
corr['price']['city-mpg']
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['engine-size'], df['price'])
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.title("Engine Size vs Price")
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='city-mpg', y='price', data=df)
plt.show()
```

```python
sns.scatterplot(x='horsepower', y='price', hue='body-style', data=df)
plt.show()
```
